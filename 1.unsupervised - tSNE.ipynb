## t-SNE

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn import datasets
from sklearn import manifold
```

```python
%matplotlib inline
```

```python
data = datasets.fetch_openml(
    'mnist_784',
    version=1,
    return_X_y=True)
```

```python
pixel_values, targets = data
targets = targets.astype(int)
```

```python
#pixel_values.iloc[1,:].values.reshape(28, 28)
```

```python
single_image = pixel_values.iloc[1,:].values.reshape(28, 28)
plt.imshow(single_image, cmap = 'gray')
```

```python
tsne = manifold.TSNE(n_components=2, init='random', random_state=42, learning_rate='auto')
transformed_data = tsne.fit_transform(pixel_values.iloc[:3000, :])
```

```python
transformed_data
```

```python
tsne_df = pd.DataFrame(
    np.column_stack((transformed_data, targets.iloc[:3000])),
    columns=["x", "y", "targets"]
)

# tsne_df.loc[:, "targets"] = tsne_df.targets.astype(int)
tsne_df[tsne_df["targets"]] = tsne_df.targets.astype(int)
```

```python
tsne_df
```
